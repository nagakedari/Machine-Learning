```python
import numpy as np
import pandas as pd
import math
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('bank-additional/bank-additional.csv', sep=";")
data.shape
```

```python
def eucliden_distance(row1, row2):
    distance = 0.0
    for i in range(len(row1)-1):
        distance += (row1[i]-row2[i])**2
    return math.sqrt(distance)
```

```python
data_subset = data.head(10)
data_subset
```

```python
data_subset['education'] = np.where(data_subset['education']=='basic.4y', 'Basic', data_subset['education'])
data_subset['education'] = np.where(data_subset['education']=='basic.6y', 'Basic', data_subset['education'])
data_subset['education'] = np.where(data_subset['education']=='basic.9y', 'Basic', data_subset['education'])

data_subset['y'] = np.where(data_subset['y']=='no', 0, data_subset['y'])
data_subset['y'] = np.where(data_subset['y']=='yes', 1, data_subset['y'])
data_subset['y'] = data_subset['y'].astype('int')
data_subset['y']
```

```python
data_subset = pd.get_dummies(data_subset, columns=['job','marital','education','default','housing','loan','contact',
                                 'month','day_of_week','poutcome'], drop_first=True)
data_subset
```

```python
data_matrix = data_subset.to_numpy()
data_matrix
row0 = data_matrix[0]
distance = []
for row in data_matrix:
    distance.append(eucliden_distance(row0, row))
distance
```

```python
def findKNearestNeibours(train, test_row, num_neighbors):
    distance = list()
    for train_row in train:
        dist = eucliden_distance(test_row, train_row)
        distance.append((train_row, dist))
    distance.sort(key=lambda tup: tup[1])
    nearest_neighbors = list()
    for i in range(num_neighbors):
        nearest_neighbors.append(distance[i][0])
    return nearest_neighbors
```

```python
neighbors = findKNearestNeibours(data_matrix, data_matrix[0], 3)
neighbors
```

```python
def predict_classification(train, test_row, num_neighbors):
    neighbors = findKNearestNeibours(train,test_row,num_neighbors)
    output_values = [row[-1] for i in neighbors]
    predict = max(set(output_values), key=output_values.count)
    return predict
```

```python
prediction = predict_classification(data_matrix, data_matrix[0], 3)
print('Expected %d, Got %d.' % (data_matrix[0][-1], prediction))
```

```python
X = data_subset.drop('y', axis=1)
y = data_subset['y']
print(y)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1)

predictions = list()
for row in X_test.to_numpy():
    predictions.append(predict_classification(X_train.to_numpy(), row, 3))

predictions
```
