```python
mean = data['GRE Score'].mean()
mode = data['GRE Score'].mode()
median = data['GRE Score'].median()

sns.distplot(data['GRE Score'], kde=False,)

plt.figure(figsize=(10,5)) # set the figsize
plt.hist(data['GRE Score'], bins=100, color='lightblue')

plt.axvline(mean, color='green')
plt.axvline(median, color='purple')
plt.axvline(mode[0], color='red')
plt.axvline(mode[1], color='orange')

plt.xlabel('GRE Score')
plt.ylabel('Frequency')
# plt.legend()

plt.show()
```

```python
# covariance
data.cov()
# If we look at the relation b/w GRE score and TOEFL score the covariance is 58 that indicates its a positive releation
# but is 58 a strong relation or weak relation. The covariance b/w GRE score and University Rating is 8, again positive
# relationship but we dont know how strong is the relation. The TOEFL scale is different vs the university rating that
# is why need to normalize to remove the effect of scale
```

```python
data.corr()
# correlation indicates the strogness of the releationship 1 is very strong
# here the correlation value for (GRE Score , TOEFL) is 0.835, similarly (GRE Score, University Rating) is 0.66 which indicates strong relation
```

```python
sns.pairplot(data, kind='reg'); #reg refers to regression
```

`Covariance and Correaltion tend to capture the idea of linearity. Linearity means that they tend to be high when there is a linear relation b/w them. When one is high the other is high in a linear sense (not exponential)`

```python
data.drop('Serial No.', axis=1,inplace=True)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
sns.heatmap(data.corr(), ax=ax, annot=True, linewidths=0.05, fmt='.2f', cmap='magma')
plt.show()

# dark color shows less correlation, lighter correlation means strong relationship
# we can say that the TOEFL score has strong relation with all the variables
```
