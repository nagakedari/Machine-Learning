```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"customer": [101,102,103,104],
                    "category": ["cat2", "cat2", "cat1", "cat3"],
                    "important": ['yes', 'no', 'no', 'yes'],
                    "sales": [123,234,345,456]
                   }, index= [0,1,2,3])
df2 = pd.DataFrame({"customer": [101,101,103,104],
                    "color": ["yellow", "green", "green", "blue"],
                    "distance": [12,9,44,21],
                    "sales": [123,213,456,765]
                   }, index=[4,5,6,7])
```

```python
pd.concat([df1, df2], axis=0, sort=False)
```

```python
pd.concat([df1, df2], axis=1, sort=True)
```

```python
pd.concat([df1, df2], axis=0, sort=True)
```

```python
df1 = pd.DataFrame({"customer": [101,102,103,105],
                    "category": ["cat2", "cat2", "cat1", "cat3"],
                    "important": ['yes', 'no', 'no', 'yes'],
                    "sales": [123,234,345,456]
                   }, index= [0,1,2,3])
df2 = pd.DataFrame({"customer": [101,101,103,104],
                    "color": ["yellow", "green", "green", "blue"],
                    "distance": [12,9,44,21],
                    "sales": [123,213,456,765]
                   }, index=[4,5,6,7])
pd.merge(df1, df2, how='outer', on='customer') # outer means union, its similar to sql outer join on is based on the common column
```

```python
# pd.merge(df1,df2, how='inner', on='customer') # inner means intersection of data frames,again similat to inner join in sql
pd.merge(df1, df2, on='customer')
```

```python
pd.merge(df1,df2, how='left', on='customer')
```

```python
pd.merge(df1,df2, how='right', on='customer')
```

```python
# join does exact the same way that a merge do, but it will combine based index value instead of columns
df3 = pd.DataFrame({'Q1':[101,102,103],
                    'Q2':[201,202,203]
                   },index=['I0','I1','I2'])
df4 = pd.DataFrame({'Q3':[301,302,303],
                    'Q4':[401,402,403]
                   },index=['I0','I2','I3'])
print(df3.join(df4, how='outer'))
print(df4.join(df3, how='outer'))
```

```python
df3.join(df4, how='inner')
```

```python
df3.join(df4, how='left')
```

```python
df3.join(df4, how='right')
```
