```python
# uber_data.drop([1155],inplace=True)
uber_data['start_dt'] = uber_data['START_DATE*'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %H:%M'))
uber_data.head()
```

```python
uber_data.info
```

```python
uber_data.dtypes
```

```python
uber_data['strt_month'] = uber_data['start_dt'].dt.month
uber_data['strt_day'] = uber_data['start_dt'].dt.day
uber_data['cal_month'] = uber_data['start_dt'].apply(lambda x: datetime.strftime(x, '%b'))
uber_data.head()
```

```python
uber_data.groupby(['cal_month', 'START*'])['cal_month'].value_counts()
```

```python
uber_data.groupby(['cal_month', 'START*'])['cal_month'].count().sort_values(ascending=False)
```

```python
uber_data.groupby(['cal_month'])['MILES*'].max().sort_values(ascending=False)
```
