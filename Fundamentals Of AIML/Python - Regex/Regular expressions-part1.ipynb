```python
import re
```

```python
text_to_search = '''
abcdefghijklmnopqrstuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890
123abc

Hello HelloHello

MetaCharacters (Need to be escaped):
. ^ $ * + ? ( ) [ ] \ | { }

utexas.edu

321-555-4321
123.555.1234

daniel-mitchell@utexas.edu

Mr. Jhonson
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
'''
```

```python
pattern = re.compile(r'abc') # r in the beginning means the raw string, which means no special treatments for special characters,
# instead they will be part of search string

matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat) # output: <re.Match object; span=(1, 4), match='abc'> string 'abc' found in position 1 to 4 (exclusive) from the string
 #<re.Match object; span=(69, 72), match='abc'>
```

```python
pattern = re.compile(r'.') # '.' means all characters except new line

matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'\.') # '\.' means finding '.'

matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
    print(mat.start(), mat.end(), mat.group(), mat.span()) # start and end gives the start and end positions
```

```python
pattern = re.compile(r'\d') # '\d' means finding digits

matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'\D') # '\.' means finding  non digits

matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'\d\w') # '\w' means finding numbers or string characters or underscore

matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'\d\s') # '\s' means finding whitespace

matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'\d\S') # '\S' means finding non-whitespace

matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'Hello') 
matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'Hello\b') # \b finding word boundary
matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'\bHello\b') 
matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'\BHello\b') # \B finding non Word boundary
matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'\b\d') # '-' (dash) considered as word boundary
matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'^\s') # ^ beginning of the string
matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```

```python
pattern = re.compile(r'$\s') # $ end of the string
matches = pattern.finditer(text_to_search)
for mat in matches:
    print(mat)
```
