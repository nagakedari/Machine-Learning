```python
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
#How many years of data is provided in the dataset? And what is the starting and ending year?

print(honey_data['year'].nunique())
print(honey_data['year'].min())
print(honey_data['year'].max())
```

```python
# Which State has seen the highest volume in production, and in which year?
honey_data[honey_data['totalprod']==honey_data['totalprod'].max()]
```

```python
#What is the average yield per colony, for each year?

honey_data[['year', 'yieldpercol']].groupby('year').mean()
```

```python
#Is there a correlation between any 2 Numeric variables? Test for the same using Visual techniques

sns.pairplot(honey_data);
```

```python
honey_data.corr() # to know how much degree the variables correleate each other. it range from -1 to +1
```

```python
sns.heatmap(honey_data.corr(), annot=True)
```

```python
# What is the general Production trend from 1998 to 2012? Describe visually
sns.jointplot(honey_data['year'], honey_data['totalprod'])
```

```python
# What is the general Production trend from 1998 to 2012? Describe visually
sns.pointplot(honey_data['year'], honey_data['totalprod'])
```

```python
sns.pointplot(honey_data['state'], honey_data['totalprod'])
```

```python
g= sns.FacetGrid(honey_data, col='state',col_wrap=5, height=3)
g.map(plt.plot, 'year', 'totalprod', marker=".")
```

```python
# Is there a linear relationship between the Number of Colonies & Value in Production? Check at an overall level, 
#at the state and year levels as well
sns.lmplot(x='numcol', y='totalprod', data=honey_data);
```

```python
sns.lmplot(x='numcol', y='prodvalue', data=honey_data, hue='year');
```

```python
#Check the distribution of total production across each year using boxplots
sns.boxplot(honey_data['year'], honey_data['totalprod'])
```

```python
 sns.barplot(honey_data['year'], honey_data['totalprod'])
```
