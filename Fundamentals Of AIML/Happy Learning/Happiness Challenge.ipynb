```python
columns = happinessData.columns
columns.shape
columns = np.delete(columns, np.where(columns == 'Score'), axis=0)
columns = np.delete(columns, np.where(columns == 'Continent'), axis=0)
for col in columns:
    sns.scatterplot(x=col, y='Score', data=happinessData)
    plt.show()
```

```python
sns.barplot(x='Continent',y='Score', data=happinessData.groupby('Continent').mean().reset_index().sort_values(by='Score'))
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(happinessData.corr(), annot=True)
plt.show()
```
